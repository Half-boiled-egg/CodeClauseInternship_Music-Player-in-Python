```python
from tkinter import *
from tkinter import filedialog
from pygame import mixer
from PIL import Image, ImageTk, ImageSequence
import os

root = Tk()
root.title("New Music Player")
root.geometry("350x600+200+10")
root.configure(background="sky blue")
root.resizable(False, False)

mixer.init()
logo = PhotoImage(file="logo.png")
root.iconphoto(False,logo)

gif_path = "musc.gif"
gif = Image.open(gif_path)
frames = [ImageTk.PhotoImage(frame) for frame in ImageSequence.Iterator(gif)]

def update(ind):
    frame = frames[ind]
    label.config(image=frame)
    root.after(40, update, (ind + 1) % len(frames))

label = Label(root)
label.pack(fill = Y)
root.after(0, update, 0)

def AddMusic():
    path = filedialog.askdirectory()
    if path:
        songs = os.listdir(path)
        for song in songs:
            if song.endswith('.mp3'):
                play_list.insert(END, song)

def PlayMusic():
    music_index = play_list.curselection()
    if music_index:
        music_name = play_list.get(music_index)
        print("Playing:", music_name)
        mixer.music.load(music_name)
        mixer.music.play()

lower_frame = Frame(root, bg="light pink", width=350, height=400)
lower_frame.place(x=0, y=300)


Menu = Frame(lower_frame, bg="sky blue", width=350, height=100)
Menu.place(x=0, y=200)

frame_music = Frame(Menu, bd=2, bg="light pink")
frame_music.place(x=0, y=25, width=350, height=99)

Play_Button = PhotoImage(file="play.png")
Button(lower_frame, image=Play_Button, bg="light pink", height=30, width=30, bd=0, command=PlayMusic).place(x=150, y=20)

Stop_Button = PhotoImage(file="stop.png")
Button(lower_frame, image=Stop_Button, bg="light pink", height=30, width=30, bd=0, command=mixer.music.stop).place(x=210, y=20)

Pause_Button = PhotoImage(file="pause.png")
Button(lower_frame, image=Pause_Button, bg="light pink", height=30, width=30, bd=0, command=mixer.music.pause).place(x=90, y=20)

Button(Menu, text="Browse Music", width=50, height=1, fg="black", bg="yellow", command=AddMusic).place(x=0, y=0)

scroll_bar = Scrollbar(frame_music)
scroll_bar.pack(side=RIGHT, fill=Y)

play_list = Listbox(frame_music, width=300, font=("Times new roman", 10), bg="light green", fg="black",
                    selectbackground="grey", cursor="hand2", bd=0, yscrollcommand=scroll_bar.set)
play_list.pack(side=RIGHT, fill=BOTH)

scroll_bar.config(command=play_list.yview)

root.mainloop()
```
